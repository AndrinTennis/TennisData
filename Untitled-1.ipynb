Load Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# To display all columns in the dataframe
pd.set_option('display.max_columns', None)
```

Load Data from ATP Tennismatches between 1993 and 2024

```python
dataframes=[]

for x in range(1993,2024):
    x=pd.read_csv(f'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_{x}.csv')
    dataframes.append(x)
df=pd.concat(dataframes,ignore_index=True)
```

```python
# Print all columns
df.columns
```

Change column's (tourney_date) datatype into date dtype + add a column (data) to the dataframe


```python
df['tourney_date']=pd.to_datetime(df['tourney_date'],format='%Y%m%d')
df['year'] = df['tourney_date'].dt.year
```

```python
temp = df[(df['best_of'] == 3) & (df['year'] > 1994)]
temp=temp.dropna(subset=['minutes', 'year'])

# Calculate the median match lengths per year
median_lengths = temp.groupby('year')['minutes'].mean().reset_index()


plt.figure(figsize=(14,5))
# Plot the median match lengths
sns.lineplot(data=median_lengths, x='year', y='minutes', marker='o', color='blue', label='Median Match Length')

# Regression line
sns.regplot(data=temp, x='year', y='minutes', scatter=False, label='Regression Line')

# Adding titles and labels
plt.title('Median Match Length in Minutes Over the last 30 Years')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Match Length in Minutes', fontsize=12)
plt.legend()
plt.show()

slope, intercept = np.polyfit(temp['year'], temp['minutes'], 1)
slope = round(slope, 2)
print(f"The rate of increase per year is {slope*60} seconds/year.")
temp = df[(df['best_of'] == 3) & (df['year'] > 1994)]
temp=temp.dropna(subset=['minutes', 'year'])

# Calculate the median match lengths per year
median_lengths = temp.groupby('year')['minutes'].mean().reset_index()


plt.figure(figsize=(14,5))
# Plot the median match lengths
sns.lineplot(data=median_lengths, x='year', y='minutes', marker='o', color='blue', label='Median Match Length')

# Regression line
sns.regplot(data=temp, x='year', y='minutes', scatter=False, label='Regression Line')

# Adding titles and labels
plt.title('Median Match Length in Minutes Over the last 30 Years')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Match Length in Minutes', fontsize=12)
plt.legend()

plt.savefig(r'C:\Users\andri\Documents\TennisData\MedianMatchLength.png')
plt.show()

slope, intercept = np.polyfit(temp['year'], temp['minutes'], 1)
slope = round(slope, 2)
print(f"The rate of increase per year is {slope*60} seconds/year.")


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

winners_df = df[df['winner_ht'].notna()]
winners_stats = winners_df.groupby('year').agg({'winner_ht': 'mean', 'w_ace': 'mean'}).reset_index()


# Create a figure and a single subplot
fig, ax1 = plt.subplots(figsize=(14, 5))

color = 'tab:blue'
# Plotting the average height of winners on the primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Height', color=color)
sns.lineplot(data=winners_stats, x='year', y='winner_ht', marker='o', color=color, ax=ax1)
ax1.tick_params(axis='y', labelcolor=color)

# Create another y-axis for the total number of aces, sharing the same x-axis
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('Total Aces', color=color)  # we already handled the x-label with ax1
sns.lineplot(data=winners_stats, x='year', y='w_ace', marker='o', color=color, ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)

# Optional: Add a title
plt.title('Average Height and Total Number of Aces by Winners Over the Years')

plt.show()
```
